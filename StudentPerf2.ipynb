```python
df=pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.drop(['lunch','test preparation course'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['parental level of education']=df['parental level of education'].fillna('No degree')
```

```python
df.head()
```

```python
df['race/ethnicity']=df['race/ethnicity'].map({
    'group A':'Asian',
    'group B':'African',
    'group C':'Afro-Asian',
    'group D':'American',
    'group E':'Europian'
})
```

```python
df.head()
```

```python
import seaborn as sns
```

```python
ax=sns.countplot(x='gender',hue='test preparation course',palette='Set1',data=df)
ax.set(title='Gender wise corse completting statistics',xlabel='Gender',ylabel='Number')
plt.show()
```

```python
ax=sns.countplot(x='gender',hue='race/ethnicity',palette='Set2',data=df)
ax.set(title='Gender Vs Student group',xlabel='Gender',ylabel='Count')
plt.show()
```

```python
interval=(0,40,50,75,100)
categories=['Failed','Seacond Class','First Class','Distinction']
df['math_cut']=pd.cut(df['mathscore'],interval,labels=categories)
```

```python
ax=sns.countplot(x='math_cut',palette='Set3',data=df)
ax.set(title='Categorical distribution of student mathscore',xlabel='Status',ylabel='Count')
```

```python
interval=(0,40,50,75,100)
categories=['Failed','Second Class','First Class','Distinction']
df['read_cut']=pd.cut(df['reading score'],interval,labels=categories)
```
